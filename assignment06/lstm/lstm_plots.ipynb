```python
%matplotlib notebook
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_acc = np.load("acc_test.npy")
train_acc = np.load("acc_training.npy")
mean_test_acc = np.load("mean_acc_test.npy")
mean_train_acc = np.load("mean_acc_training.npy")
```

```python
print(mean_test_acc.T.shape)
plt.plot(mean_test_acc.T)
plt.title("Mean fraction correct on test set")
plt.xlabel("Epochs")
plt.ylabel("Mean fraction correct on test set")
plt.legend(['0','1','2','3','4','5','6','7'],loc='best',fancybox=True, framealpha=0.5)
```

```python
#calculating the standard error of the mean (= standard deviation / sqrt(10))    
print(np.std(test_acc,axis=1).T)
temp= np.std(test_acc,axis=1).T / np.sqrt(10)
plt.plot(temp)
plt.title("Standard error of the mean")
plt.xlabel("Epochs")
plt.ylabel("Standard error of the mean")
plt.legend(['0','1','2','3','4','5','6','7'],loc='best',fancybox=True, framealpha=0.5)
```
